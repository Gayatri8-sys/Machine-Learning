```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_scaled = scaler.fit_transform(df.select_dtypes(include='number'))
df_scaled=pd.DataFrame(df_scaled,columns=df.select_dtypes(include='number').columns)
```

```python
data=pd.concat([pd.DataFrame(df_scaled), df.select_dtypes(exclude='number')], axis=1)
data
```

```python
for col in data.select_dtypes(exclude='number').columns:
  print(f"{col} has {data[col].unique()} unique values \n")
  print(f"{col} has {data[col].nunique()} unique values \n")
  print("-------------------------------------------------")
```

```python
# race sex income ----> one hot encoding

# workclass education marital status, occupation relationship native country -------------> label encoding


from sklearn.preprocessing import LabelEncoder , OneHotEncoder
le=LabelEncoder()


col1=['workclass','education','marital_status','occupation','relationship','native_country']
col2=['race','sex','income']

for col in col1:
  data[col]=le.fit_transform(data[col])

for col in col2:
  oh=OneHotEncoder(sparse_output=False)
  encoded=oh.fit_transform(data[[col]])
  encoded_df = pd.DataFrame(encoded, columns=oh.get_feature_names_out([col]))

  data = data.drop(col, axis=1)
  data = pd.concat([data.reset_index(drop=True), encoded_df.reset_index(drop=True)], axis=1)
```

```python
# â€¢	Discuss the pros and cons of One-Hot Encoding and Label Encoding.
```

One-Hot Encoding
What it does:
Converts each category value into a new binary column (0 or 1).

Pros:

Handles nominal categorical data well (no order assumed).

Avoids implying any ordinal relationship between categories.

Works well with algorithms that cannot handle categorical variables directly (e.g., linear regression, logistic regression).

Cons:

Can cause high dimensionality if the number of categories is large (curse of dimensionality).

Increases memory and computation cost.

Can lead to sparse matrices if many categories exist.





---



Label Encoding
What it does:
Assigns each category an integer value (0, 1, 2, ...).

Pros:

Simple and efficient (only one column added).

Good for ordinal categorical data where order matters (e.g., "low", "medium", "high").

Uses less memory compared to one-hot encoding.

Cons:

Implies an ordinal relationship even when there is none, which can mislead some algorithms.

Some ML models might interpret the numeric labels as ordered values, causing bias.

Not suitable for nominal categorical features without natural order.

```python
data['net_capitals']=data['capital_gain']-data['capital_loss']
data.drop(columns=['capital_gain','capital_loss'], inplace=True)
```

```python
# age groups

bins=[0,25,45,60,100]
labels=['Young', "Middle",'Senior','Old']
data['age_group']=pd.cut(data['age'], bins=bins, labels=labels)
data.drop(columns=['age'], inplace=True)
```

```python
data
```

```python
# pd.cut is a pandas function used to segment and sort continuous numerical data into discrete bins or intervals.
# To convert continuous variables into categorical bins.
# Useful for creating groups like age ranges, income brackets, etc.
# Makes it easier for models to capture nonlinear relationships.
```

```python
data['log_capital_gain'] = np.log1p(data['net_capitals'])
data.drop(columns=['net_capitals'], inplace=True)
```

Feature Selection

```python
# isolation Forest ---- unsupervised algo, used for outliers removal
```

```python
from sklearn.ensemble import IsolationForest

num_cols = data.select_dtypes(include=['int64', 'float64']).columns
X = data[num_cols]

# Fit Isolation Forest
iso = IsolationForest(contamination=0.02, random_state=42)
outliers = iso.fit_predict(X)
```

```python
outliers    # -1 means outlier, 1 means normal
```

```python
data['outlier'] = outliers

data = data[data['outlier'] == 1].drop('outlier', axis=1)
```
