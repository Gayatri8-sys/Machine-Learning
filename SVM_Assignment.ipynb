```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
data=pd.read_csv("/content/mushroom.csv")
```

```python
data
```

```python
features=[['cap_surface','odor','habitat','ring_type']]
```

```python
data.describe()
```

```python
data.dtypes
```

```python
sns.pairplot(data)
```

```python
from sklearn.preprocessing import LabelEncoder

encoded_df=data.copy()

le=LabelEncoder()
for col in data.columns:
  if encoded_df[col].dtype=='object':
    encoded_df[col]=le.fit_transform(encoded_df[col])
```

```python
encoded_df
```

```python
X=encoded_df.drop(columns=['Unnamed: 0', 'class'])
Y=encoded_df['class']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
# hypertuning for multiple parameters
# ---RBF---------
from sklearn.model_selection import GridSearchCV
clf=SVC()
param_grid=[{'kernel':['rbf'],'gamma':[50,5,10,0.5], 'C':[15,14,13,12,11,10,0.1,0.001]}]
gsv=GridSearchCV(clf,param_grid,cv=10)
gsv.fit(X_train, Y_train)
```

```python
gsv.best_params_, gsv.best_score_
```

```python
clf=SVC(kernel='rbf', C=15, gamma=50)

clf.fit(X_train, Y_train)
Y_pred=clf.predict(X_test)

from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, precision_score
rbf_acc=accuracy_score(Y_test, Y_pred)*100
print("Accuracy=", rbf_acc)
print("confusion matrix:", confusion_matrix(Y_test, Y_pred))
print("f1 score:", f1_score(Y_test, Y_pred))
print("precision score:", precision_score(Y_test, Y_pred))
```

```python
cm=confusion_matrix(Y_test, Y_pred)
sns.heatmap(cm, annot=True)
plt.show()
```

```python
# ----LINEAR--------

clf=SVC()
param_grid={
    'kernel': ['poly'],
    'gamma': [0.1, 0.5],
    'C': [1, 10]
}
gsv=GridSearchCV(clf, param_grid, cv=10)
gsv.fit(X_train, Y_train)
```

```python
gsv.best_params_, gsv.best_score_
```

```python
clf=SVC(kernel='linear', C=1 ,gamma=0.1 )
clf.fit(X_train, Y_train)
Y_pred=clf.predict(X_test)
linear_acc=accuracy_score(Y_test, Y_pred)*100
print("Accuracy=", linear_acc)
print("confusion matrix:", confusion_matrix(Y_test, Y_pred))
print("f1 score:", f1_score(Y_test, Y_pred))
print("precision score:", precision_score(Y_test, Y_pred))
```

```python
# ------POLY----------

from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
clf=SVC()
param_grid={
    'kernel': ['poly'],
    'gamma': [0.1, 0.5],
    'C': [1, 10]
}
gsv=GridSearchCV(clf, param_grid, cv=10)
gsv.fit(X_train, Y_train)
```

```python
clf=SVC(kernel='poly', C= 1, gamma= 0.1)
clf.fit(X_train, Y_train)
Y_pred=clf.predict(X_test)
poly_acc=accuracy_score(Y_test, Y_pred)*100
print("Accuracy: ", poly_acc)
print("confusion matrix:", confusion_matrix(Y_test, Y_pred))
print("f1 score:", f1_score(Y_test, Y_pred))
print("precision score:", precision_score(Y_test, Y_pred))
```

```python
kernels = ['rbf', 'linear', 'poly']
accuracies=[rbf_acc, linear_acc, poly_acc]
plt.bar(kernels, accuracies, color=['skyblue', 'lightgreen', 'salmon'])
plt.xlabel('kernels')
plt.ylabel('accuracy')
plt.show()
```
