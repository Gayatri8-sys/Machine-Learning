```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/anime.csv")
```

```python
df
```

```python
data=df[['anime_id', 'name', 'genre', 'rating']]
```

```python
data.set_index('name', inplace=True)
```

```python
data
```

```python
data['genre'].fillna("", inplace=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf=TfidfVectorizer(stop_words='english')
tfidf_matrix=tfidf.fit_transform(data['genre'])
```

```python
# compute cosine similarity matrix

from sklearn.metrics.pairwise import cosine_similarity
cosine_sim=cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim_df=pd.DataFrame(cosine_sim)
cosine_sim_df.index=data.index
cosine_sim_df.columns=data.index
cosine_sim_df
```

```python
np.fill_diagonal(cosine_sim_df.values, 0)
```

```python
cosine_sim_df
```

```python
cosine_sim_df.idxmax(axis=1)
```

```python
def recommend(anime_name, top_n=5):
    if anime_name not in cosine_sim_df.index:
        return "Anime not found."

    # Get similarity scores for the input anime
    sim_scores = cosine_sim_df.loc[anime_name].drop(anime_name)

    # Sort by similarity score, descending
    top_animes = sim_scores.sort_values(ascending=False).head(top_n).index

    # Return the top recommended animes with genre & rating
    return data.loc[top_animes, ['genre', 'rating']].reset_index()
```

```python
print(recommend('Wind: A Breath of Heart (TV)',5))
```

Can you explain the difference between user-based and item-based collaborative filtering?

User-Based Collaborative Filtering:
Recommends items based on similar users.
If User A likes Anime X and Y, and User B is similar to A, then recommend X or Y to B.
Finds user-user similarity.


Item-Based Collaborative Filtering:
Recommends items similar to what a user has already liked.
If a user liked Anime X, recommend items similar to X.
Finds item-item similarity.


---



What is collaborative filtering, and how does it work?

Collaborative Filtering is a recommendation technique that suggests items (like movies, anime, or products) based on the preferences of many users — assuming that similar users will like similar things.

It uses a user-item matrix (who rated what) to find patterns like:

Users who liked similar items

Items liked by similar users

It doesn’t need content info (like genre or description) — just user behavior.
